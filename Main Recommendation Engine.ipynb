```python
import numpy as np
import pandas as pd
```

```python
anime = pd.read_csv('animes.csv')
```

```python
anime.head()
```

```python
anime.shape
```

```python
anime = anime[anime.episodes != -1]
anime = anime[anime.score != -1]
anime = anime[anime.type != "Unknown"]
```

```python
anime=anime.reset_index(drop=True)
```

```python
anime.shape
```

```python
anime = anime[['mal_id','title','synopsis','genres','type']]
```

```python
anime.head()
```

```python
anime.isnull().sum()
```

```python
anime.dropna(inplace=True)
```

```python
anime.duplicated().sum()
```

```python
anime.isnull().sum()
```

```python
anime=anime.reset_index(drop=True)
```

```python
anime.shape
```

```python
anime['synopsis'][1]
```

```python
anime['synopsis'] = anime['synopsis'].apply(lambda x:x.split())
```

```python
anime.head()
```

```python
 anime['genres'] = anime['genres'].apply(lambda x:x.split())
```

```python
anime['tags'] = anime['synopsis'] + anime['genres']
```

```python
anime.head()
```

```python
new_df = anime[['mal_id' , 'title' , 'tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.shape
```

```python
#Text Vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=7000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:11]
```

```python
def recommend(anime):
    anime_index = new_df[new_df['title'] == anime].index[0]
    distances = similarity[anime_index]
    anime_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:11]
    
    for i in anime_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Naruto')
```

```python
recommend('Death Note')
```
